## Modelling - build pipeline.

```python
from sklearn.model_selection import train_test_split

X = df.drop(['default', 'LOAN_ID'], axis=1)
y = df['default']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
X.info()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_selector as selector
```

```python
num_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
cat_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
df.info()
```

```python
num_features = df.select_dtypes(include=['int64', 'float64']).drop(['PD_RISK_RATING', 'default'], axis=1)
cat_features = df.select_dtypes(include=['object']).drop(['LOAN_ID'], axis=1)

numeric_cols = df.dtypes.apply(lambda x: x.kind in 'bifc').reset_index(drop=True).loc[lambda x: x == True].index
cat_cols = (df.dtypes == 'object').reset_index(drop=True).loc[lambda x: x == True].index


preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transform,  selector(dtype_exclude="object")),
        ('cat', cat_transform, selector(dtype_include="object"))])
```

```python
## Add SMOTE to deal with imbalanced nature of defaults

from numpy import mean
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import classification_report



over = SMOTE(sampling_strategy=0.1)
under = RandomUnderSampler(sampling_strategy=0.5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                     ('over', over), 
                     ('under', under),
                     ('classifier', LogisticRegression(solver='lbfgs'))])
```

```python
# evaluate pipeline
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(pipeline, X, y, scoring='roc_auc', cv=cv, n_jobs=-1)
print('Mean ROC AUC: %.3f' % mean(scores))
```

```python
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
```

```python
print("model score: %.3f" % f1_score(y_test, y_pred, average='weighted'))
```

```python
print(classification_report(y_test, y_pred))
```
